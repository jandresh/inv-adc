```python
import json
 
# Opening JSON file
f = open('gui/src/authors_50.json')
 
# returns JSON object as
# a dictionary
data = json.load(f)
 
# Iterating through the json
# list
for i in data:
    print(i)
 
# Closing file
f.close()
```

```python
nodes = "{nodes: ["
for node in data[0]['authors_list']:
    nodes += "{id:\""
    nodes += node
    nodes += "\"}, "
nodes += "], "
print(nodes)
```

```python
authors_len = len(data[0]['authors_list'])
print(f"authors_len: {authors_len}")
links = "links: ["
for i in range(authors_len):
    for j in range(authors_len):
        if (data[0]['array'][i][j]>0 and i!=j):
            links += '{'
            links += f'"source": "{data[0]["authors_list"][i]}", "target": "{data[0]["authors_list"][j]}"'
            links += '}, '
links += "]}"
print(f'{nodes}{links}')
```

```python
import requests
import time
from datetime import datetime
import json
def post_json_request(url, obj):
    return requests.post(url, json=obj).json()

container = "orchestrator"

def create_logger(microservice):
    create_state_db = post_json_request(
        'http://localhost:5001/mongo-db-create',
        {"db_name" : "app_state"}
    )
    create_state_coll = post_json_request(
        'http://localhost:5001/mongo-coll-create',
        {
            "db_name" : "app_state",
            "coll_name" : f"{container}_{microservice}"
        }
    )
    print(f"{create_state_db['exit']}:{create_state_coll['exit']}")
    return create_state_db["exit"] == 0 and create_state_coll["exit"] == 0
def state_logger(microservice, state):
    print(dict(
            db_name= "app_state",
            coll_name=f"{container}_{microservice}",
            document=dict(
                datetime=datetime.now().isoformat(),
                state=state
            )
        ))
    return post_json_request(
        'http://localhost:5001/mongo-doc-insert',
        dict(
            db_name= "app_state",
            coll_name=f"{container}_{microservice}",
            document=dict(
                datetime=datetime.now().isoformat(),
                state=state
            )
        )
    )

maxdocs = 500
microservice_name = "pipeline7"
while (not create_logger(microservice_name)):
    time.sleep(10)
    print("no db connection")
state_logger(
    microservice_name,
    json.dumps(obj=dict(
        stage=0,
        description="Pipeline 7 start",
        data=dict(pattern=0, success=False)
    ))
)
```
