```python
! pip install scipy
! pip install numpy
import requests
import networkx as nx

def post_json_request(url, obj):
    return requests.post(url, json=obj).json()
```

```python
authors = post_json_request(
    "http://localhost:5001/mongo-doc-list",
    {
        "db_name": "fluidattacks",
        "coll_name": "authors#BreastCancer2#global"
    }
)
```

```python
authors_list = [f'{author["author"]} {" ".join(author["related"])}' for author in authors]
authors_list
```

```python
G = nx.parse_adjlist(authors_list, nodetype=str)
G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
nx.clustering(G)
```

```python
import matplotlib.pyplot as plt
nx.draw(G)
```
